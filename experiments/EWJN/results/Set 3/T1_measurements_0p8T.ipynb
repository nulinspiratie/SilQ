Next step is to tune ESR, which is proving to be quite difficult.
There are some frequencies with high contrast, but these frequencies change from scan to scan.
As an example, below two scans are shown with identical settings, both with a 40 MHz adiabatic frequency window.

```python
# 40 MHz window
data = qc.load_data('2016-12-04/#014_adiabatic_sweep_12-04-55')
qc.MatPlot(data.contrast);
```

```python
# 40 MHz window, same meaurement settings, performed right after
data = qc.load_data('2016-12-04/#015_adiabatic_sweep_12-16-19')
qc.MatPlot(data.contrast);
```

It is clear that the results differ significantly, although the reason why is not clear.
One other interesting point is that in the second scan, the frequency range with high contrast is around 200 MHz, much higher that the 40 MHz window.
This suggests that the actual resonance frequency could also be shifting, which would indicate a changing magnetic field.

As a check, we perform a scan at the same frequency (22.376 GHz), with 450 samples.
The results below show that the contrast switches between high and low.
I do not know why the high contrast is significantly lower than before, perhaps the energy landscape shifted resulting in a less optimal voltage point.
